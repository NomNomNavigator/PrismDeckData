# Clean and create csv for movie tag

```python
# id | user_id | movie_id | tag | create_ts
```

```python
import pandas as pd
```

```python
df = pd.read_csv('raw_data/tags.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df[df['tag'].isnull()]
```

```python
# 17 missing tags, we decided to drop these rows
df = df.dropna()
```

```python
# When loading into database, discovered that some tags are very long. Example below.
df.iloc[[9402]]
```

```python
# Decided to remove tags longer than 100 characters
length_condition = df['tag'].str.len() <= 100
```

```python
df = df[length_condition]
```

```python
df
```

```python
df.index.name='id'
```

```python
# Renaming columns to match database
df = df.rename(columns={"userId": "user_id", "movieId": "movie_id", "timestamp": "create_ts"})
```

```python
# Change genres to lowercase
df['tag'] = df['tag'].str.lower()
```
