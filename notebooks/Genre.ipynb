# Create a csv to load into the genres table

```python
import pandas as pd
```

```python
df = pd.read_csv('DB_models/DB_genres.csv', index_col = 0)
```

```python
df
```

```python
new_rows = []
for _, row in df.iterrows():
    movie_id = row['movieId']
    genres = row['genre'].split('|')
    for genre in genres:
        new_rows.append({'movieId': movie_id, 'genre': genre})
```

```python
new_df = pd.DataFrame(new_rows)
```

```python
new_df
```

```python
new_df.index.name='id'
```

```python
new_df = new_df.rename(columns={"movieId": "movie_id"})
```

```python
new_df
```

```python
new_df.to_csv('DB_models/DB_genres.csv')
```
